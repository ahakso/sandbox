# NB summary:
This nb 
1. Uses selenium to launch a browser
1. the user then carries out manual steps
    1. The user logs in with their own credentials, and navigates to a page with a region dropdown
    1. The user selects a region, and hides all non-selected countries
1. selenium finds the buttons with the selected countries
1. IDs that are consistent with the data as delivered are available in the app. selenium pulls the country names and iso values associated with each selected button
1. user repeats for each region

Shortcoming: WW-MAJOR is not available in the web app. We may just want to ignore that one

Dissuaded because web app disagrees with documentation. Favor web app since the html references the same IDs they give us for country iso

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
wait = WebDriverWait(driver, 3)

# Navigate to page
driver.get("https://apptopia.com")

# then manually login, click a region and show only selected
```
