```python
w = 24
h = 95
theta = 90
def th(theta):
    h1 = w*np.sin(theta)
    h2 = h*np.cos(theta)
    return h1+h2
    
```

```python
theta = np.linspace(0,np.pi/2,100)
thv = [th(x) for x in theta]
```

```python
90/np.max(theta)
```

```python
plt.plot(theta*(90/np.max(theta)),thv)
plt.gca().set(ylim=(90,99))
```
